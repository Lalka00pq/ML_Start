```python
import onnxruntime as onnx
import numpy as np
import torch
import torch.nn as nn
import torchvision.models as models
```

### Перевод модели ResNet18

```python

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

state_dict = torch.load(
    r"C:\Users\User\Desktop\ML_projects\ML_Start_Caisar\ML_Start\src\4-е задание\best_model_ResNet18_2.pth", map_location=device).state_dict()


model = models.resnet18()
model.fc = nn.Sequential(
    nn.Linear(model.fc.in_features, 512),
    nn.ReLU(),
    nn.Dropout(),
    nn.Linear(512, 2)
)
model.load_state_dict(state_dict=state_dict, strict=False)
model.to(device)
model.eval()
```

```python
import torch.onnx as onnx_torch
dummy_input = torch.randn(1, 3, 224, 224, requires_grad=True).to(device)
```

```python
output = model(dummy_input)
```

```python
onnx_torch.export(model, dummy_input, "model_resnet18_best_loss.onnx")
```

### Перевод модели RegNet

```python
state_dict_regnet = torch.load(
    r'C:\Users\User\Desktop\ML_projects\ML_Start_Caisar\ML_Start\src\5-е задание\best_model_RegNetY_400MF_1.pth', map_location=device).state_dict()
```

```python
model_regnet = models.regnet_y_400mf()
model_regnet.fc = nn.Sequential(
    nn.Linear(model_regnet.fc.in_features, 512),
    nn.ReLU(),
    nn.Dropout(),
    nn.Linear(512, 2)
)
model_regnet.load_state_dict(state_dict=state_dict_regnet, strict=False)
model_regnet.to(device)
model_regnet.eval()
```

```python
dummy_input_regnet = torch.randn(1, 3, 224, 224, requires_grad=True).to(device) 
output_regnet = model_regnet(dummy_input_regnet)
torch.onnx.export(model_regnet, dummy_input_regnet, "model_regnet_best_loss.onnx")
```
